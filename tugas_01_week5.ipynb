```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

import os
from google.cloud.sql.connector import connector


os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "C:/Users/sage-outrider-418915-a6610c425b1b.json"
```

```python
# Dummy connection parameters
DB_NAME = "postgres"
DB_USER = "postgres"
DB_PASSWORD = "admin"
DB_HOST = "localhost"
DB_PORT = "5432"

try:
    # Establish a connection to the PostgreSQL database
    connection = psycopg2.connect(
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD,
        host=DB_HOST,
        port=DB_PORT
    )

    # Create a cursor object using the connection
    cursor = connection.cursor()

    # Define your SQL query to select data from a table
    query = "select * from tugas_01 where gender not like '%E%' and age not like '%U%'"

    # Execute the query
    cursor.execute(query)
    
    # Fetch all rows from the result set
    rows = cursor.fetchall()

    # Get the column names from the cursor description
    columns = [desc[0] for desc in cursor.description]

    # Create a pandas DataFrame from the fetched rows and column names
    df = pd.DataFrame(rows, columns=columns)

    columns_to_remove_quotes = ["customer","age","gender","zipcodeOri","merchant","zipMerchant","category"]

    for column in columns_to_remove_quotes:
        df[column] = df[column].str.replace("'", "")


    # Close the cursor and connection
    cursor.close()
    connection.close()

except (Exception, psycopg2.Error) as error:
    print("Error while connecting to PostgreSQL:", error)
finally:
    # Close the connection if it's still open
    if connection:
        connection.close()
```
